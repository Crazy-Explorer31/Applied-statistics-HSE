```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from scipy import special
from tqdm import tqdm
```

```python
q = 0.95
sigma = 1
delta = 0.2 # неснижаемая погрешность (погрешность измерения)
n_arange = np.linspace(5, 1000, 100)

interval_semiwidth = stats.norm.ppf((1 + q) / 2) * (sigma / np.sqrt(n_arange))

data = pd.DataFrame({'x': n_arange, 'y': interval_semiwidth})

plt.figure(figsize=(14, 6))

sns.lineplot(data=data, x='x', y='y', marker='o', markersize=5, label='Полуширина доверительного интервала')
plt.hlines(y=delta, xmin=0, xmax=1000, color='red', label='Неснижаемая погрешность')
plt.legend()
plt.title(f"График неснижаемой погрешности")
plt.xlabel("Размер выборки")
plt.ylabel("Погрешность")

plt.tight_layout()
plt.show();
```
