```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Оценки квантилей

```python
N = 10**3
quantiles = [np.zeros(1000) for _ in np.arange(2, 5.1, 0.5)]
for i, size_degree in enumerate(np.arange(2, 5.1, 0.5)):
    size = 10 ** size_degree
    for j in range(1000):
        sample = np.random.uniform(0, 1, int(size))
        quantile = np.quantile(sample, 0.01, method="nearest")
        quantiles[i][j] = quantile
```

```python
[np.mean(quantiles_sized) for quantiles_sized in quantiles]
```

```python
quantiles_vars = [np.var(quantiles_sized) for quantiles_sized in quantiles]
```

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(quantiles[0], fill=True, alpha=0.4, label='100 наблюдений')
sns.kdeplot(quantiles[1], fill=True, alpha=0.4, label='300 наблюдений')
sns.kdeplot(quantiles[2], fill=True, alpha=0.4, label='1000 наблюдений')
sns.kdeplot(quantiles[3], fill=True, alpha=0.4, label='3000 наблюдений')
# sns.kdeplot(quantiles[4], fill=True, alpha=0.4, label='10000 наблюдений')
# sns.kdeplot(quantiles[5], fill=True, alpha=0.4, label='30000 наблюдений')
# sns.kdeplot(quantiles[6], fill=True, alpha=0.4, label='100000 наблюдений')


plt.title('0.01 квантили')
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
N = 10**3
medianes = [np.zeros(1000) for _ in np.arange(2, 5.1, 0.5)]
for i, size_degree in enumerate(np.arange(2, 5.1, 0.5)):
    size = 10 ** size_degree
    for j in range(1000):
        sample = np.random.uniform(0, 1, int(size))
        median = np.quantile(sample, 0.5, method="nearest")
        medianes[i][j] = median
```

```python
[np.mean(medianes_sized) for medianes_sized in medianes]
```

```python
medianes_vars = [np.var(medianes_sized) for medianes_sized in medianes]
```

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(medianes[0], fill=True, alpha=0.4, label='100 наблюдений')
sns.kdeplot(medianes[1], fill=True, alpha=0.4, label='300 наблюдений')
sns.kdeplot(medianes[2], fill=True, alpha=0.4, label='1000 наблюдений')
sns.kdeplot(medianes[3], fill=True, alpha=0.4, label='3000 наблюдений')
# sns.kdeplot(medianes[4], fill=True, alpha=0.4, label='10000 наблюдений')
# sns.kdeplot(medianes[5], fill=True, alpha=0.4, label='30000 наблюдений')
# sns.kdeplot(medianes[6], fill=True, alpha=0.4, label='100000 наблюдений')


plt.title('Медианы')
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
N = 10**3
expectations = [np.zeros(1000) for _ in np.arange(2, 5.1, 0.5)]
for i, size_degree in enumerate(np.arange(2, 5.1, 0.5)):
    size = 10 ** size_degree
    for j in range(1000):
        sample = np.random.uniform(0, 1, int(size))
        expectation = np.mean(sample)
        expectations[i][j] = expectation
```

```python
[np.mean(expectations_sized) for expectations_sized in expectations]
```

```python
expectations_vars = [np.var(expectations_sized) for expectations_sized in expectations]
```

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(expectations[0], fill=True, alpha=0.4, label='100 наблюдений')
sns.kdeplot(expectations[1], fill=True, alpha=0.4, label='300 наблюдений')
sns.kdeplot(expectations[2], fill=True, alpha=0.4, label='1000 наблюдений')
sns.kdeplot(expectations[3], fill=True, alpha=0.4, label='3000 наблюдений')
# sns.kdeplot(expectations[4], fill=True, alpha=0.4, label='10000 наблюдений')
# sns.kdeplot(expectations[5], fill=True, alpha=0.4, label='30000 наблюдений')
# sns.kdeplot(expectations[6], fill=True, alpha=0.4, label='100000 наблюдений')


plt.title('Мат ожидания')
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Зависимость дисперсии от размера выборки

```python
np.arange(2, 5.1, 0.5)
```

```python
data_q = pd.DataFrame({'x': np.arange(2, 5.1, 0.5), 'y': quantiles_vars})
data_m = pd.DataFrame({'x': np.arange(2, 5.1, 0.5), 'y': medianes_vars})
data_e = pd.DataFrame({'x': np.arange(2, 5.1, 0.5), 'y': expectations_vars})

plt.figure(figsize=(10, 6))
sns.lineplot(data=data_q, x='x', y='y', marker='o', markersize=8, label='Квантили 0.01')
sns.lineplot(data=data_m, x='x', y='y', marker='o', markersize=8, label='Медианы')
sns.lineplot(data=data_e, x='x', y='y', marker='o', markersize=8, label='Мат ожидания')
plt.title('Зависимость дисперсии от размера выборки')
plt.xlabel("Логарифм размера выборки")
plt.ylabel("Оценка дисперсии")
plt.legend()
plt.show()
```
